```python
import math
def perception():
  x1 = float(input("Enter x1: "))
  x2 = float(input("Enter x2: "))

  w1 = float(input("Enter w1: "))
  w2 = float(input("Enter w2: "))

  b = float(input("Enter b: "))

#weighted
  z =(x1*w1)+(x2*w2)+b
  print("Weighted sum = ", z)
  

  #sigmoid function
  s = 1/(1 + math.exp(-z))
  return s

result = perception()
print("Perception Output =" , result)
```
